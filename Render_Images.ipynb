### Render images

```python

#load dataset
#os.chdir(r'C:\Users\babet\Documents\Studi\Master Thesis\Table_detection')
os.chdir(r'E:\Babette\MasterThesis')
df = pd.read_pickle('gs_renderable.pkl')

#set path for images
#os.chdir(r'C:\Users\babet\Documents\Studi\Master Thesis\Table_detection\Bilder')
os.chdir(r'E:\Babette\MasterThesis\IMG')

#Set path to wkhtmltoimage
path_wkhtmltoimage = r'C:\Users\babet\Documents\Studi\Master Thesis\Table_detection\Resources\wkhtmltox-0.12.5-1.mxe-cross-win64\wkhtmltox\bin\wkhtmltoimage.exe'
config = imgkit.config(wkhtmltoimage=path_wkhtmltoimage)
```

```python
#Render images with archive

renderError=[]

for x in df.index:
    if df.label.iloc[x]== "RELATION":
        os.chdir(r'E:\Babette\MasterThesis\IMG\RELATION')   
    elif df.label.iloc[x]== "MATRIX":
        os.chdir(r'E:\Babette\MasterThesis\IMG\MATRIX') 
    elif df.label.iloc[x]== "ENTITY":
        os.chdir(r'E:\Babette\MasterThesis\IMG\ENTITY') 
    elif df.label.iloc[x]== "OTHER":        
        os.chdir(r'E:\Babette\MasterThesis\IMG\OTHER') 
        
    path = r'id_' + str(df.id.iloc[x])+ '.jpg'
    html = df.renderCodeImage.iloc[x]
    
    if df.archivePath.iloc[x] != 'NAN':
         try:
            imgkit.from_string(html, path, config=config)
            print(df.id.iloc[x])
            renderError.append(0)
        except:
            print("error occurred:")
            print(df.id.iloc[x])
            renderError.append(1)
    else:
        renderError.append("NAN")
```

```python
df["renderErrorArchive"]= renderError #improve!
```

### Create Train, Validation & Test sample

```python
#Sample data into train/test (80/20) 

rel_id=[]
mat_id=[]
ent_id=[]
oth_id=[]

for x in df.index:
    if df.archivePath.iloc[x]!= "NAN":
        if df.label.iloc[x]== "RELATION":
            rel_id.append(df.id.iloc[x])
        if df.label.iloc[x]== "MATRIX":
            mat_id.append(df.id.iloc[x])
        if df.label.iloc[x]== "ENTITY":
            ent_id.append(df.id.iloc[x])
        if df.label.iloc[x]== "OTHER":
            oth_id.append(df.id.iloc[x])
            
        
```

```python
y= round(len(rel_id)/100*20) 
rel_id_test=(random.sample(rel_id, y))

y= round(len(mat_id)/100*20) 
mat_id_test=(random.sample(mat_id, y))

y= round(len(ent_id)/100*20) 
ent_id_test=(random.sample(ent_id, y))

y= round(len(oth_id)/100*20) 
oth_id_test=(random.sample(oth_id, y))
```

```python
#relocate images create RANDOM TEST SET

for x in rel_id_test:
    path= "E:\Babette\MasterThesis\IMG\RELATION\id_"+str(x)+".jpg"
    
    try:
        img = Image.open(path)
        
        new_path = "E:\Babette\MasterThesis\IMG\Test\RELATION\id_"+str(x)+".jpg"
        img.save(new_path, "JPEG")
        
        os.remove(path)
    except:
        print("failed:", x)
        
        

      
```
